```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_algo = pd.read_csv('./baguncado.csv')
df_algo['x'] = df_algo['x'].astype(int)
df_algo['tempo'] = df_algo['tempo'].astype(float)
df_algo
```

```python
tarefa_2 = df_algo[df_algo.tarefa == 'cargadetrabalho2']
sns.lineplot(x=tarefa_2.x, y=tarefa_2.tempo)
sns.scatterplot(x=tarefa_2.x, y=tarefa_2.tempo)

plt.xlabel('tamanho da entrada')
plt.ylabel('tempode execucao')
plt.title('carga de trabalho2 O(1),caso bagunçado')
```

```python

tarefa_1 = df_algo[df_algo.tarefa == 'cargatrabalho1']
sns.lineplot(x=tarefa_1.x, y=tarefa_1.tempo, label='Cargadetrabalho1', color='blue')
sns.scatterplot(x=tarefa_1.x, y=tarefa_1.tempo, color='blue')

tarefa_4 = df_algo[df_algo.tarefa == 'bubblesort']
sns.lineplot(x=tarefa_4.x, y=tarefa_4.tempo, label='Bubblesort', color = 'black')
sns.scatterplot(x=tarefa_4.x, y=tarefa_4.tempo, color='black')

tarefa_7 = df_algo[df_algo.tarefa == 'insertosort']
sns.lineplot(x = tarefa_7.x, y=tarefa_7.tempo, label = 'insertosort',color = 'green')
sns.scatterplot(x = tarefa_7.x, y=tarefa_7.tempo,color = 'green')

tarefa_11 = df_algo[df_algo.tarefa == 'selectionsort']
sns.lineplot(x = tarefa_11.x, y = tarefa_11.tempo, label = 'selectionsort',color= 'gray')
sns.scatterplot(x=tarefa_11.x, y = tarefa_11.tempo, color = 'gray')

tarefa_12 = df_algo[df_algo.tarefa == 'shellsort']
sns.lineplot(x = tarefa_12.x, y = tarefa_12.tempo,label = 'shellsort', color = 'pink')
sns.scatterplot(x = tarefa_12.x, y = tarefa_12.tempo ,color = 'pink')

plt.xlabel('tamanho da entrada')
plt.ylabel('tempo em microsegundos')
plt.title('grafico de tempo de execução O(n^2),caso bagunçado')
```

```python
tarefa_3 = df_algo[df_algo.tarefa == 'cargadetrabalho3']
sns.lineplot(x = tarefa_3.x,y = tarefa_3.tempo,label = 'Carga de trabalho3',color = 'orange')
sns.scatterplot(x = tarefa_3.x, y = tarefa_3.tempo,color = 'orange')


tarefa_5 = df_algo[df_algo.tarefa == 'bucketsort']
sns.lineplot(x = tarefa_5.x, y = tarefa_5.tempo,label='bucketsort',color='black')
sns.scatterplot(x = tarefa_5.x, y = tarefa_5.tempo,color = 'black')

tarefa_6 = df_algo[df_algo.tarefa == 'countingsort']
sns.lineplot(x = tarefa_6.x,y = tarefa_6.tempo,label = 'countingsort', color = 'blue')
sns.scatterplot(x = tarefa_6.x,  y = tarefa_6.tempo, color = 'blue')


tarefa_10 = df_algo[df_algo.tarefa == 'radixsort']
sns.lineplot(x = tarefa_10.x,y = tarefa_10.tempo, label ='radixsort', color = 'gray')
sns.scatterplot(x= tarefa_10.x,y = tarefa_10.tempo, color = 'gray')


plt.xlabel('tamanho da entrada')
plt.ylabel('tempo de execucao')
plt.title('grafico da carga de trabalho O(n),caso bagunçado')
```

```python
tarefa_8 = df_algo[df_algo.tarefa == 'quicksort']
sns.lineplot(x = tarefa_8.x, y = tarefa_8.tempo,label = 'quicksort',color = 'black')
sns.scatterplot(x = tarefa_8.x, y = tarefa_8.tempo, color = 'black')

tarefa_9 = df_algo[df_algo.tarefa == 'mergesort']
sns.lineplot(x = tarefa_9.x , y = tarefa_9.tempo, label = 'mergesort',color = 'green')
sns.scatterplot(x = tarefa_9.x , y = tarefa_9.tempo, color = 'green')



plt.xlabel('tamanho da entrada')
plt.ylabel('tempo de execução')
plt.title('grafico da carga de trabalho O(nlogn),caso bagunçado')
```
